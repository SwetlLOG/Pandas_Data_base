1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома Если реновации дома не было, то в новом признаке поставьте 0

```python
df["delta_renovated"] = df['yr_renovated'] - df['yr_built']

df.head()
```

```python
df.loc[df["delta_renovated"]<0,"delta_renovated"] = 0
df.head()
```

```python
df['sells_years'] = pd.to_datetime(df['date'], format="%Y")
df.head()
```

```python
df['sells_month'] = pd.to_datetime(df['date'], format="%m")
df.head()
```

#### Условие 3: 3 задача
Составьте несколько сводных таблиц

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей








```python
df.head()
```

```python
df.groupby('bedrooms').agg({'price':'mean'}).sort_values('price')
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('grade').agg({'price':['min', 'mean', 'max']})
```

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index='waterfront',
               columns='view',
               values= 'bathrooms',
               aggfunc='count',
               fill_value=0)
```

3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df.pivot_table(index='floors',
               columns='bedrooms',
               values= 'bathrooms',
               aggfunc='count',
               fill_value=0)
```

Больше всего одноэтажных домов с количеством спален 3. 

```python
pd.crosstab(index=df['floors'],
            columns=df['bedrooms']).max().max()
```

3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома 
и оценки дома

```python
df.groupby(['grade','condition']).agg({'price':'mean'}).unstack('grade').fillna(0)
```
